```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import csv data
airbnba = "./Resources/listings .csv" 
airbnba_df = pd.read_csv(airbnba)
airbnba_df.head()
```

```python
# delete the column neighbourhood_group
del airbnba_df['neighbourhood_group']
del airbnba_df['name']
airbnba_df.head()
```

```python
total_rows = airbnba_df.count
total_rows
```

```python
#Keep only the rows without non-NA values
airbnba_df.dropna(inplace=True)
airbnba_df.head()
```

```python
total_rows = airbnba_df.count
total_rows
```

```python
# set index for the dataframe
finalairbnba_df= airbnba_df .reset_index()
finalairbnba_df.head(10)
```
