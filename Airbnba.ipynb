```python
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

```python
#import csv data
airbnba = "./listings .csv" 
airbnba_df = pd.read_csv(airbnba)
airbnba_df.head()
```

```python
# delete the column neighbourhood_group
del airbnba_df['neighbourhood_group']
del airbnba_df['name']
airbnba_df.head()
```

```python
total_rows = airbnba_df.count
total_rows
```

```python
#Keep only the rows without non-NA values
airbnba_df.dropna(inplace=True)
airbnba_df.head()
```

```python
total_rows = airbnba_df.count
total_rows
```

```python
# set index for the dataframe
finalairbnba_df= airbnba_df .reset_index()
finalairbnba_df.head(10)
#finalairbnba_df.to_csv("cleaned_listings.csv")
```

```python
#Connect to local databaseT
rds_connection_string = f'root:{password}@localhost:3306/airbnbaustin_db'
```

```python
engine = create_engine(f'mysql+pymysql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
finalairbnba_df.to_sql(name='finalairbnba_data', con=engine, if_exists='append', index=False)
```
